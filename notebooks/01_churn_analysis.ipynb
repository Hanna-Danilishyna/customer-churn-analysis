```python
import pandas as pd
from sqlalchemy import create_engine, text

engine = create_engine("postgresql://postgres@localhost:5432/postgres")
df = pd.read_sql("SELECT * FROM customer_features", engine)
```

```python
df.groupby("churn")[[
    "total_orders",
    "total_items",
    "total_revenue",
    "avg_order_value",
    "recency"
]].mean()
```

```python
df["churn"].value_counts().plot(kind="bar")
```
